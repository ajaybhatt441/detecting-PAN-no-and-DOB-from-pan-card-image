```python
import cv2
import matplotlib.pyplot as plt
import pytesseract
import re
```

```python
image=cv2.imread('pan_card.jpg')
plt.figure(figsize=(10,10))
plt.imshow(image)
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
ret,thres=cv2.threshold(gray,100,255,0)
plt.figure(figsize=(10,10))
plt.imshow(thres)
data=pytesseract.image_to_string(thres)
pan_number=re.findall("[A-Z0-9]{10}",data)[1]
date_of_birth=re.findall('[0-9]*/[0-9]*/[0-9]*',data)
print('PAN NO : {}'.format(pan_number))
print('D.O.B : {}'.format(date_of_birth))
```
